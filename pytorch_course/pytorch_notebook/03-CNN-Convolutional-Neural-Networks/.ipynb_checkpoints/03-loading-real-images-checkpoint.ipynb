```python
transform = transforms.Compose([
    transforms.Resize(250), # resize (250px x X)
    transforms.CenterCrop(250), # crop to (250px x 250px)
    transforms.RandomHorizontalFlip(p=1), # horizontally flip with 100% chance (usually set to p=0.5)
    transforms.RandomRotation(30), # +/- 30 deg
    transforms.ToTensor()
])
im = transform(dog)
plt.imshow(np.transpose(im.numpy(), (1,2,0)))
```

```python
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]) # values usually work well
])

im = transform(dog)
plt.imshow(np.transpose(im.numpy(), (1,2,0)))
```

```python
im
```

```python
inv_normalize = transforms.Normalize(
    mean=[-0.485/0.229, -0.456/0.224, -0.406/0.225],
    std=[1/0.229, 1/0.224, 1/0.225]
)
im_inv = inv_normalize(im)
plt.figure(figsize=(12,4))
plt.imshow(np.transpose(im_inv.numpy(), (1,2,0)))
```
